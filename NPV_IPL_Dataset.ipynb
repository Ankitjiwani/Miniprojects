```python
df2 = pd.read_csv('matches.csv')
df2
```

```python
print("Total number of Matches: ",df2['id'].count())
```

```python
print("Total number of Seasons: ",df2['season'].nunique())
```

```python
print("Win by Most runs :",df2.loc[df2['win_by_runs'].max(),'winner'])
```

```python
print("Win by Most wickets :",df2.loc[df2['win_by_wickets'].max(),'winner'])
```

```python
print("Win by Minimun runs :",df2[df2['win_by_runs'].ge(1)].min()['winner'])
```

```python
print("Win by Minimun Wickets :",df2[df2['win_by_wickets'].ge(1)].min()['winner'])
```

```python
sns.countplot(x='season', data=df2)
plt.show()
print("Season 2013 has most number of matches")
```

```python
data = df2.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h');

print("Mumbai Indians is the most Successful Team")
```

```python
ss = df2['toss_winner'] == df2['winner']

print(ss.groupby(ss).size())

print("Yes, Toss winning helped in winning matches.")
```
